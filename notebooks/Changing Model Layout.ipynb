```python
from keras.models \
    import load_model, Sequential

from keras.layers \
    import BatchNormalization, Conv2D, Dense

import wandb
```

```python
model: Sequential = load_model(
    '/opt/models/O2RM'
)

layers = model.layers

for index in range(len(layers)):
    print(index, layers[index])
```

```python
layers.insert(
    9, 
    BatchNormalization()
)

layers.insert(
    5, 
    BatchNormalization()
)

layers.insert(
    3, 
    BatchNormalization()
)
```

```python
model = Sequential(layers)
model.summary()
```

```python
model.save(
    '/opt/models/O2RM',
    save_format='tf'
)
```
