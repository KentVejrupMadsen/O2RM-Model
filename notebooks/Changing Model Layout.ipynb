```python
from keras.models import load_model, Sequential
from keras.layers import BatchNormalization, Conv2D, Dense

from tensorflow import expand_dims

import wandb
```

```python
model: Sequential = load_model(
    '/opt/programming/ORM-Model/resources/Model'
)

layers = model.layers
model.summary()
```

```python
size_of_layers: int = len(layers)
last_index_of_layers: int = size_of_layers - 1

remove_at: list = list()

for layer_index in range(last_index_of_layers):
    layer = layers[
        layer_index
    ]
    
    if isinstance(
        layer,
        BatchNormalization
    ):
        remove_at.append(
            layer_index
        )

for l in reversed(
    remove_at
):
    layers.pop(
        l
    )
```

```python
rescale = layers[0]

layers.insert(
    1, 
    Conv2D(
        3, 
        3, 
        name='new_layer_1',
        input_shape=rescale.input_shape[1:],
        padding="same",
        activation='relu'
    )
)

layers.insert(
    2, 
    Conv2D(
        3, 
        3, 
        name='new_layer_2',
        input_shape=rescale.input_shape[1:],
        padding="same",
        activation='relu',
    )
)

layers.insert(
    3, 
    Conv2D(
        3, 
        3, 
        name='new_layer_3',
        input_shape=rescale.input_shape[1:],
        padding="same",
        activation='relu'
    )
)

layers.insert(
    4, 
    Conv2D(
        3, 
        3, 
        name='new_layer_4',
        input_shape=rescale.input_shape[1:],
        padding="same",
        activation='relu'
    )
)
```

```python
for i in range(
    len(
        layers
    )
):
    layer = layers[i]
    
    print(
        i, 
        layer
    )
```

```python
last_index = len(layers) - 1

layers[
    last_index - 3
]

layers[
    last_index - 2
] = Dense(
    1024
)

layers[
    last_index - 1
] = Dense(
    512
)

layers[
    last_index
] = Dense(
    10
)

model = Sequential(
    layers
)
model.summary()
```

```python
model.save(
    '/opt/models/O2RM',
    save_format='tf'
)
```
